```python
empleados = [
    {"nombre": "Carlos", "salario": 1500, "departamento": "Finanzas"},
    {"nombre": "María", "salario": 2500, "departamento": "Tecnología"},
    {"nombre": "Pedro", "salario": 1200, "departamento": "Administración"},
    {"nombre": "Ana", "salario": 3200, "departamento": "Finanzas"},
]

def filtrar_por_salario(empleados, umbral):
    return [empleado for empleado in empleados if empleado["salario"] > umbral]

empleados_filtrados = filtrar_por_salario(empleados, 2000)
print("Empleados con salario mayor a 2000:")
for emp in empleados_filtrados:
    print(emp["nombre"], emp["salario"])

def salario_por_departamento(empleados):
    salarios = {}
    for empleado in empleados:
        departamento = empleado["departamento"]
        salarios[departamento] = salarios.get(departamento, 0) + empleado["salario"]
    return salarios


salarios_departamento = salario_por_departamento(empleados)
print("Salario total por departamento:")
for depto, salario in salarios_departamento.items():
    print(f"{depto}: {salario}")

```

```python
def generar_informe_mensual(empleados, departamento):
    total_gastos = 0
    for empleado in empleados:
        if empleado['departamento'] == departamento:
            total_gastos += empleado['gastos']
    return total_gastos

empleados = [
    {"nombre": "Carlos", "gastos": 300, "departamento": "Finanzas"},
    {"nombre": "María", "gastos": 450, "departamento": "Administración"},
    {"nombre": "Ana", "gastos": 200, "departamento": "Finanzas"},
    {"nombre": "Pedro", "gastos": 500, "departamento": "Tecnología"}
]

departamentos = ["Finanzas", "Administración", "Tecnología"]

for dept in departamentos:
    informe = generar_informe_mensual(empleados, dept)
    print(f"Total de gastos en el departamento de {dept}: ${informe}")
```

```python
def filtrar_empleados_por_cargo(empleados):
    ejecutivos = []
    operativos = []
    for empleado in empleados:
        if empleado["salario"] > 3000:
            ejecutivos.append(empleado["nombre"])
        else:
            operativos.append(empleado["nombre"])
    return ejecutivos, operativos

empleados = [
    {"nombre": "Carlos", "salario": 2500},
    {"nombre": "Ana", "salario": 3500},
    {"nombre": "Pedro", "salario": 2800},
    {"nombre": "María", "salario": 4000}
]

ejecutivos, operativos = filtrar_empleados_por_cargo(empleados)

print("Empleados Ejecutivos:", ejecutivos)
print("Empleados Operativos:", operativos)
```

```python
def calcular_multa(dias_retraso, multa_por_dia=50):
    if dias_retraso > 0:
        multa_total = dias_retraso * multa_por_dia
    else:
        multa_total = 0
    return multa_total

proyectos = [
    {"nombre": "Proyecto A", "dias_retraso": 5},
    {"nombre": "Proyecto B", "dias_retraso": 0},
    {"nombre": "Proyecto C", "dias_retraso": 3}
]

for proyecto in proyectos:
    multa = calcular_multa(proyecto["dias_retraso"])
    print(f"Multa para {proyecto['nombre']}: ${multa}")
```
